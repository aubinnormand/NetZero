```python
import subprocess
import os
from pathlib import Path; import ipynbname; 
# Définir le chemin du dépôt
repo_path = ipynbname.path().parent.parent

# Fonction pull
def git_pull():
    subprocess.run(["git", "pull", "origin", "main", "--allow-unrelated-histories"])

# Fonction add/commit/push
def git_add_commit_push(message="Mise à jour du projet"):
    subprocess.run(["git", "add", "."])
    subprocess.run(["git", "commit", "-m", message], check=True)
    subprocess.run(["git", "push", "-u", "origin", "main"])

def git_update_from_remote():
    subprocess.run(["git", "fetch", "origin"])
    subprocess.run(["git", "reset", "--hard", "origin/main"])
```

```python
git_update_from_remote()
```

```python
git_pull()  # récupérer les dernières modifications
```
